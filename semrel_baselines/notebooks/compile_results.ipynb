```python
import pandas as pd
import os
import json
import glob
import os
```

```python
os.chdir('..')
```

```python
list_of_langs = ["arb", "arq", "ary", "esp", "hau", "hin", "ind", "kin", "mar", "pan", "tel", "afr", "amh"]
path_lists = []
for lang in list_of_langs:
    path = os.path.join("results", lang, f"similarity_evaluation_{lang}_test_home_results.csv")
    path_lists.append(path)
```

```python
df_results = pd.DataFrame()
for lang, path in zip(list_of_langs,path_lists):
    df = pd.read_csv(path)
    spearman_scores = df['cosine_spearman'].iloc[0]
    df_temp = pd.DataFrame({'lang': lang, 'spearman_scores': spearman_scores}, index=[0])
    df_results = pd.concat([df_results, df_temp])
    # df_results = df_results.append(df)
```

```python
df_results = df_results.round(2)
```

```python
df_results.to_csv('results/simularity_evaluation_eng_finetuned_sbert.csv', index=False)
```

```python
df_results.to_csv("results/similarity_evaluation_all_langs_test_sentence_bert_finetuned_results.csv", index=False)
```

```python
json_files = glob.glob('results/results_lexical/**/*.json', recursive=True)
```

```python
json_files
```

```python
# read the above json_files and create a dataframe with columns: lang, spearman_scores, model_name for the corresponding json file
df_results = pd.DataFrame()
for json_file in json_files:
    with open(json_file) as f:
        data = json.load(f)
        lang = data['lang_name']
        spearman_scores = data['spearman_correlation']
        # model_name = data['model_name']
        df_temp = pd.DataFrame({'lang': lang, 'spearman_correlation': spearman_scores}, index=[0])
        df_results = pd.concat([df_results, df_temp])
```

```python
df_results = df_results.round(2)
```

```python
df_results.to_csv('results/simularity_evaluation_lexical.csv', index=False)
```

```python
# pivot the dataframe to have columns: lang, model_name, spearman_scores
df_fin = df_results.pivot(index='lang', columns='model_name', values='spearman_correlation')
```

```python
# reset index_df_fin
df_fin.reset_index(inplace=True)
```

```python
# replace NaN with -
df_fin = df_fin.fillna('-')
```
